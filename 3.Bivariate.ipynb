```python
# variance inflation factor is imported from the satsmodels.stats.outliers_influence module

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a function named as calc_vif(X),X is the input variable

def calc_vif(X):
    
    #calculating VIF,DataFrame is created and stiord in ithe name vif
    vif=pd.DataFrame()
# Input column index names tabulted as variables name
    vif["variables"]=X.columns
# By uuing the for loop each and every coumn values of the input columns assigned in the name i 
# and find the vif by using x.values.
    vif["VIF"]=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    return(vif)
```

# ANAVO : Analysis of Variance

```python
# Oneway Classification
```

```python
import scipy.stats as stats

stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

```python
# pvale=0.5 There is variance between these given columns
```

```python
# Twoway Classification
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fit the ANOVA model
model = ols("dataset['salary'] ~ dataset['ssc_p'] * dataset['hsc_p']", dataset).fit()

# Perform ANOVA test
aov_table = sm.stats.anova_lm(model, typ=2)
```

```python
aov_table
```

```python
# F value is higher than p value (PR>F) There is significance variance between ssc_p and salary also hsc_p and salary
# ssc_p and hsc_p There is no sigificance variance between these two
```
